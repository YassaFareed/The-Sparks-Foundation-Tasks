Importing Libraries

```python
import keras_ocr
import matplotlib.pyplot as plt
```

Loading the pretrained weights 

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

Creating a list of images 

```python
images = [
          keras_ocr.tools.read(img) for img in ['a1.jpg',
                                                'a2.jpg'
          ]
]
```

Predictions

```python
predictions_group = pipeline.recognize(images)
```

Annotated Output from predictions

```python
fig, axs = plt.subplots(nrows = len(images), figsize=(10,20))
for ax, image, predictions in zip(axs,images,predictions_group):

  keras_ocr.tools.drawAnnotations(image=image,
                                  predictions=predictions,
                                  ax=ax
                                  ) 
```

Words present in both images 

```python
count=1
for x in predictions_group:
  print('--------{} images--------'.format(count))
  for i in x:
    print(i[0])
  count+=1
```
